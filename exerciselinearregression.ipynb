```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pd.read_csv('linear regression/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df.price,color='red')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.price)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df.price,color='red')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
plt.show()
```
