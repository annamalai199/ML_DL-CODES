```python
import pandas as pd
df=pd.read_csv("salaries.csv")
df.head()
```

```python
input=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
input
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
company=LabelEncoder()
job=LabelEncoder()
degree=LabelEncoder()
```

```python
input['company_n']=company.fit_transform(input['company'])
input['job_n']=job.fit_transform(input['job'])
input['degree_n']=degree.fit_transform(input['degree'])
input.head()
```

```python
input_n=input.drop(['company','job','degree'],axis='columns')
input_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
model.predict([[2,0,0]])
```

```python
df1=pd.read_csv("titanic.csv")
df1.head
```

```python
input1 = df1.drop(['Survived', 'PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Embarked', 'Cabin'], axis='columns')
target1=df1['Survived']
target1
```

```python
from sklearn.preprocessing import LabelEncoder
input1['Pclass_1'] = LabelEncoder().fit_transform(df1['Pclass'])
input1['Sex_1'] = LabelEncoder().fit_transform(df1['Sex'])
input1['Age_1'] = LabelEncoder().fit_transform(df1['Age'])
input1['Fare_1'] = LabelEncoder().fit_transform(df1['Fare'])
input1.head()
```

```python
input_n1=input1.drop(['Pclass','Sex','Age','Fare'],axis='columns')
input_n1
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(input_n1,target1)
```

```python
model.score(input_n1,target1)
```
