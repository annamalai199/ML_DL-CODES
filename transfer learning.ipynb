```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np

# Step 1: Define input shape
IMAGE_SHAPE = (224, 224)

# Step 2: Create a safe wrapper for the hub layer
hub_layer = hub.KerasLayer("https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/4", trainable=False)
wrapped_layer = tf.keras.layers.Lambda(lambda x: hub_layer(x))

# Step 3: Build the model using Functional API
inputs = tf.keras.Input(shape=IMAGE_SHAPE + (3,))
outputs = wrapped_layer(inputs)
model = tf.keras.Model(inputs=inputs, outputs=outputs)

# Step 4: Run prediction on a dummy image
dummy_image = np.random.rand(1, 224, 224, 3).astype(np.float32)
predictions = model(dummy_image)

# Step 5: Get the top class
top_class = tf.argmax(predictions, axis=-1).numpy()[0]
print("Prediction shape:", predictions.shape)
print("Top predicted class index:", top_class)
```

```python
import tensorflow as tf
import tensorflow_hub as hub

# Define number of flower classes
num_of_flowers = 5

# Input layer
inputs = tf.keras.Input(shape=(224, 224, 3))

# Define the feature extractor layer FIRST (do NOT pass input_shape here)
feature_extractor_layer = hub.KerasLayer(
    "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4",
    trainable=False
)

# Pass input through feature extractor
x = feature_extractor_layer(inputs)  # ✅ This works

# Classification head
outputs = tf.keras.layers.Dense(num_of_flowers, activation='softmax')(x)

# Build model
model = tf.keras.Model(inputs=inputs, outputs=outputs)

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Show model summary
model.summary()
```
