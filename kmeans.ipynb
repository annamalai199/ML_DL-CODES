```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import os
os.environ["OMP_NUM_THREADS"] = "1"
import pandas as pd
df=pd.read_csv("income.csv")
df.head()
```

```python
%matplotlib inline
plt.scatter(df['Age'],df['Income($)'])
plt.show()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predict=km.fit_predict(df[['Age','Income($)']])
y_predict
```

```python
df['cluster']=y_predict
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1[['Income($)']],color='green')
plt.scatter(df2.Age,df2[['Income($)']],color='red')
plt.scatter(df3.Age,df3[['Income($)']],color='yellow')
plt.show()
```

```python
scalar=MinMaxScaler()
scalar.fit(df[['Income($)']])
df['Income($)']=scalar.transform(df[['Income($)']])
df
```

```python
scalar.fit(df[['Age']])
df['Age']=scalar.transform(df[['Age']])
df
```

```python
km2=KMeans(n_clusters=3)
y_predict1=km2.fit_predict(df[['Age','Income($)']])
y_predict1
```

```python
df['cluster12']=y_predict1
df
```

```python
df1=df[df.cluster12==0]
df2=df[df.cluster12==1]
df3=df[df.cluster12==2]
plt.scatter(df1.Age,df1[['Income($)']],color='green')
plt.scatter(df2.Age,df2[['Income($)']],color='red')
plt.scatter(df3.Age,df3[['Income($)']],color='yellow')

plt.show()
```

```python
km2.cluster_centers_
```

```python
df1=df[df.cluster12==0]
df2=df[df.cluster12==1]
df3=df[df.cluster12==2]
plt.scatter(df1.Age,df1[['Income($)']],color='green')
plt.scatter(df2.Age,df2[['Income($)']],color='red')
plt.scatter(df3.Age,df3[['Income($)']],color='yellow')
plt.scatter(km2.cluster_centers_[:,0],km2.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.show()
```

```python

k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit_predict(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_rng,sse)
plt.show()
```
