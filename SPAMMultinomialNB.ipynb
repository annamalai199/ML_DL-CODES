```python
import pandas as pd
df=pd.read_csv("spam.csv")
df.head()
```

```python
  df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
email=['hey mohan can we get together to watch football game tomorrow']
e_count=v.transform(email)
model.predict(e_count)
```

```python
x_test_count=v.transform(x_test)
model.score(x_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('vectorizer', CountVectorizer()),       # Step 1: Convert text to numbers
    ('nb', MultinomialNB())     # Step 2: Train the model
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.score(x_test,y_test)
```
